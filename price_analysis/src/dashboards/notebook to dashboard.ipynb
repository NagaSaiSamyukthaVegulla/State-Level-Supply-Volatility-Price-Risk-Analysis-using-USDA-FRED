```python
%sql 
with top_states as (
  select state_name , year , total_production , avg_price_received,
  rank() over(partition by year order by total_production desc) as rnk
  from priceriskanalysis.gold.usda_corn_gold_state_analysis
  where year = (
    select max(year) from priceriskanalysis.gold.usda_corn_gold_state_analysis
  )
)
select state_name, total_production,avg_price_received from top_states
where rnk <=5
order by total_production desc;
```

year over year price change

```python
%sql 
with price_cte as(
  select year,avg_price_received,
  lag(avg_price_received) over(order by year) as prev_year_price 
  from priceriskanalysis.gold.usda_corn_gold
)
select year, 
case 
   when prev_year_price is null then null 
else round(avg_price_received - prev_year_price,2) end as price_change
from price_cte
order by year;
```
